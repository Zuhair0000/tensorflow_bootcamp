Create Tensors with tf.Variable

```python
changable_tensor = tf.Variable([10, 7])
unchangable_tensor = tf.constant([10, 7])

changable_tensor, unchangable_tensor
```

change elements

```python
changable_tensor[0].assign(7)
changable_tensor
```

```python
unchangable_tensor[0].assign(7)
unchangable_tensor
```

create two random (but the same) tensors

```python
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(3, 2))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3, 2))
random_2
```

Suffle the order of elements in a tensor

```python
not_shuffled = tf.constant([[10, 7],
                           [3, 4],
                           [2, 5]])
not_shuffled.ndim
```

```python
not_shuffled
```

```python
tf.random.set_seed(42)
not_shuffled = tf.random.shuffle(not_shuffled, seed=42)
not_shuffled
```

Numpy arrays to tensors

Getting information from tensors

* Shape
* Size
* Rank
* Axis

```python
rank_4_tensor = tf.zeros(shape=(2, 3, 4, 5))
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

Get various attributes of tensor

```python
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions of every element:", rank_4_tensor.ndim)
print("Shapae of tensot:", rank_4_tensor.shape)
print("Elemnst along the 0 axis:", rank_4_tensor.shape[0])
print("Elemnst along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements:", tf.size(rank_4_tensor).numpy())
```
